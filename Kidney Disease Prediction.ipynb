```python
df = pd.read_csv('pre_processed_df_kidney_disease.csv')
df.info()
```

```python
df = df.drop(['Unnamed: 0'],axis=1)
df.head()
```

```python
df.classification.value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('classification',axis=1)
y = df['classification']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from imblearn.over_sampling import ADASYN
adasyn = ADASYN()
X_resampled, y_resampled = adasyn.fit_resample(X_train,y_train)
```

```python
y_resampled.value_counts()
```

```python
from sklearn.svm import LinearSVC
lin_svc = LinearSVC()
lin_svc.fit(X_resampled,y_resampled)
```

```python
from sklearn.metrics import accuracy_score, classification_report
y_preds = lin_svc.predict(X_test )
accuracy = accuracy_score(y_test,y_preds)
print(f"Accuracy: {accuracy}")
print(classification_report(y_test, y_preds))
```

```python
from sklearn.model_selection import cross_val_score
lin_svc_cv_score = cross_val_score(estimator=lin_svc,X=X_resampled,y=y_resampled,scoring='accuracy',cv=5)
lin_svc_cv_score.mean()
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression(max_iter=10000)
log_reg.fit(X_resampled,y_resampled)
```

```python
from sklearn.metrics import accuracy_score, classification_report
y_preds = log_reg.predict(X_train)
accuracy = accuracy_score(y_test,y_preds)
print(f"Accuracy: {accuracy}")
print(classification_report(y_test, y_preds))
```
